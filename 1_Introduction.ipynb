<span style="font-size:35px; font-weight:bold;">`Introduction`</span><br>
### `About the Dataset`
This project is based on Hospital Management dataset. It has 5 main CSV files:
#### `patients.csv` 
- Patient demographics, contact details, registration info, and insurance data
#### `doctors.csv` 
- Doctor profiles with specializations, experience, and contact information
#### `appointments.csv` 
- Appointment dates, times, visit reasons, and statuses
#### `treatments.csv` 
- Treatment types, descriptions, dates, and associated costs
#### `billing.csv` 
- Billing amounts, payment methods, and status linked to treatments.

### `Project Objective`
The goal of this project is to perform Exploratory Data Analysis (EDA) on hospital data and find key trends and patterns. Using visualizations, it examines billing, payments, doctor and branch performance, patient demographics, and treatment trends. The insights from these visuals help understand hospital operations, highlight areas of improvement, and support better decision-making for patient care and revenue management.
